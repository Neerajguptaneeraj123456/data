```python
df.head(2)
import pandas as pd
import random
from datetime import datetime, timedelta

# Load your file (adjust the filename if needed)


# Generate dates between Jan 1, 2024 and Dec 31, 2025
start_date = datetime(2022-23, 1, 1)
end_date = datetime(2024, 12, 31)

admission_dates = []
discharge_dates = []

for _ in range(len(df)):
    admission = start_date + timedelta(days=random.randint(0, (end_date - start_date).days - 15))
    discharge = admission + timedelta(days=random.randint(2, 15))

    admission_dates.append(admission.strftime("%d/%m/%Y"))
    discharge_dates.append(discharge.strftime("%d/%m/%Y"))

# Update DataFrame
df["Date of Admission"] = admission_dates
df["Discharge Date"] = discharge_dates

# Save as new CSV
df.to_csv("healthcare_dataset_indian_doctors_updated_2024_2025.csv", index=False)
print("✅ File saved as 'healthcare_dataset_indian_doctors_updated_2024_2025.csv'")

df.head()
# df.info()
```
